```python
%matplotlib notebook
# This lets jupyter plot in-line
#Can also use: %matplotlib inline
import matplotlib.pyplot as plt
import random
import numpy as np

fig1 = plt.figure(1)
#Create Random List
rand_list = [random.random() for _ in range(20)]
_ = plt.hist(rand_list)
plt.show()
#If I don't use _ = then the array returns in-line
 
#Mean
avg = np.mean(rand_list)
print("Mean:\t%.4f\t(average)" %avg)
#Median
print("Median:\t%.4f\t(datapoint nearest middle)" %np.median(rand_list))
```

```python
#STD
std = np.std(rand_list)
print("Sigma:\t%.4f\t(standard deviation)" %std)
#Variance
var = np.var(rand_list)
print("Var:\t%.4f\t(Variance)" %var)
print("Var=std**2?:  %s" %(var == std**2))
print("Normal distribution and standard deviations:")
from IPython.display import Image
Image("Std_dev.jpg",width=350,height=350)
#PDF
#CDF
```

```python
#Now let's overlay our histogram plot with a normal distribution
plt.close(fig1)
fig2 = plt.figure(2)
ax1=fig2.add_subplot(111)
_ = plt.hist(rand_list)
#setup normal distribution
from scipy.stats import norm
x = np.linspace(avg-3*std, avg+3*std,100)
ax2 = fig2.add_subplot(111, sharex=ax1, frameon=False)
ax2.yaxis.tick_right()
plt.plot(x, norm.pdf(x,avg,std),'r')
plt.show()
```

```python
'''
PDF and CDF
PDF = Probability distribution function
A distribution that shows the probability at any X value
CDF = Cumulative distribution function
Distribution shows the probability summed up to that X value
'''
plt.close(fig2)
dx=0.01
x = np.arange(-2,2,dx)
y = np.exp(-x**2)
# normalize to 1
y /= (y*dx).sum()
cdfy = np.cumsum(y*dx)
fig3 = plt.figure(3)
plt.plot(x,y,'r',label="PDF")
plt.plot(x,cdfy,'k--',label="CDF")
_=plt.legend(loc=2)
```
